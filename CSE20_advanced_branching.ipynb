````python
# insurance premium
color = "gold"
has_racing_kit = True
model = "Prius"

base_premium = 100
if model != "Prius":
    if color == "red":
        base_premium += 50
    if has_racing_kit:
        base_premium += 75
else:
    if color == "gold":
        base_premium += 100
    else:
        base_premium -= 15
    
    
print("Total car insurance premium: $", base_premium)
````

#  `elif` Statements

- `elif` statements are used to control the execution of more than 2 mutually exclusive blocks of code.
- The Python interpreter will start evaluating with the `if` statement and then continue in order until the condition an `elif` is `True` or it reaches the `else` statement 

```python
if [condition_A]: 
    print("'[condition_A]' was True")
elif [condition_B]:
    print("'[condition_A]' was False and '[condition_B]' was True")
else:
    print("Both '[condition_A]' and '[condition_B]' were False")
```

# `elif` Statements

````python
speed_limit = 65
current_speed = 66

if current_speed < speed_limit:
    print("You are traveling slower than the speed limit")
elif current_speed == speed_limit:
    print("You are traveling at the speed limit")
else:
    print("You are traveling faster than the speed limit")
````

# `elif` Statements

````python
digit = 2

if digit == 0:
    print("zero")
elif digit == 1:
    print("one")
elif digit == 2:
    print("two")
elif digit == 3:
    print("three")
else:
    print("Sorry, I can only count to three...")
````

### The Department of Motor Vehicles (DMV)

- Every year car owners pay a registration fee to drive their cars
- The DMV has two types of registration **Standard**(\\$150) and **Planned Non-Operation (PNO)**(\\$25)
- There is a late fee that increases as a function of time and is different for both types:
    - 0 days late: Standard=\\$0, PNO=\\$0
    - 1-7 days late: Standard=\\$25, PNO=\\$5
    - 7-14 days late: Standard=\\$54, PNO=\\$12
    - 14-31 days late: Standard=\\$156, PNO=\\$35
    - \> 31 days late: Standard=\\$212, PNO=(Not allowed)
