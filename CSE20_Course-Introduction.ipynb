# Late Assignments

There will be 5 bi-weekly assignments over the course of the quarter. Assignments will involve programming. 
Assignments are usually due every other Sunday at 11:59 PM, and are turned in electronically. Rather than 
approve extensions on a case-by-case basis, each student has 2 "grace days," each of which may be used, no 
explanation necessary, to extend the due date of a (single) assignment by 24 hours. Two grace days cannot be 
used on the same assignment, and half grace days are not permitted. Use your grace days wisely!
