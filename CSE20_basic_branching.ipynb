# Review: `bool` the Boolean Type

- A `bool` type variable can take one of two values:`True` or `False`
- They are used in conjunction with binary and unary logical operators to perform branching (also called conditional control flow)
- A common naming convention for `bool` variables is name them in pseudo question form: `is_valid_input` or `is_greater`


# Review: `bool` the Boolean Type

````python
user_input = int(input("Please enter a number:"))

is_greater_than_3 = user_input > 3

print(user_input, "is greater than 3?", is_greater_than_3)
````

# Binary Logical Operators

| Symbol |          Meaning          | Precedence |             Example             |
|:------:|:-------------------------:|:----------:|:-------------------------------:|
|   `>`  |        Greater Than       |    Lower   |      `5 > 2 # equals True`      |
|   `<`  |         Less Than         |    Lower   |      `5 < 2 # equals False`     |
|  `>=`  | Greater Than  or Equal To |    Lower   |      `5 >= 2 # equals True`     |
|  `<=`  |   Less Than or Equal To   |    Lower   |     `5 <= 2 # equals False`     |
|  `==`  |          Equal To         |    Lower   |     `5 == 2 # equals False`     |
|  `!=`  |        Not Equal To       |    Lower   |      `5 != 2 # equals True`     |
|  `and` |        Logical And        | Even Lower | `True and False # equals False` |
|  `or`  |         Logical Or        | Even Lower |  `True or False # equals True`  |


# Logical `and` & `or` Truth Table

| Operand 1 | Operand 2 |  `and`  |   `or`  |
|:---------:|:---------:|:-------:|:-------:|
|   `True`  |   `True`  |  `True` |  `True` |
|   `True`  |  `False`  | `False` |  `True` |
|  `False`  |   `True`  | `False` |  `True` |
|  `False`  |  `False`  | `False` | `False` |

# Unary Logical Operators

| Symbol |          Meaning          | Precedence |             Example             |
|:------:|:-------------------------:|:----------:|:-------------------------------:|
|  `not` |      Logical Negation     |    Lower   |    `not True # equals False`    |

### Logical `not` Truth Table

| Operand |  `not`  |
|:-------:|:-------:|
|  `True` | `False` |
| `False` |  `True` |

# Branching 

- It is often the case that while writing a program, we would like to execute certain portions of our code only under certain conditions.
- This is analogous to the branching we see in flow charts

# `if` Statements

- We can use an `if` statement to model this kind of behavior
- You can nest code inside an `if` statement and the code within the `if` statement is only executed if the condition is `True`
- Code is "scoped" within an `if` statement by being written below it and being **indented 4 spaces**
- Though tabs are *technically* acceptable, 4 spaces is the standard adopted by the community, and so it should be what you use

# `if` Statements

```python
if [condition]:   # note that the line ends with a ':'
    # code to execute if condition is true
    # note also this is indented by 4 spaces
    print("Looks like '[condition]' was True")
```

# `if` Statements

````python
customer_rate = 150.00

customer_age = int(input("Please enter your age:"))

if customer_age > 55:
    customer_rate /= 2 # equivalently, `customer_rate = customer_rate / 2`
    
print("Please pay $", customer_rate)
print("Have a great day!")
````

# `if`/`else` Statements 

- We can express this kind of logic using an `if` and `else` statement.
- An `if/else` statement has is an `if` statement with an added `else` after
- Code that is executed conditionally is indented 4 spaces
- `else` statements **cannot** stand on their own, they need an `if` 

```python
if [condition]:   # note that the line ends with a ':'
    # code to execute if condition is true
    # note also this is indented by 4 spaces
    print("Looks like '[condition]' was True")
else: # this line also ends with a ':'
    print("Looks like '[condition]' was False")
```

# `if`/`else` Statements 

````python
user_prompt = "Have you tried turning it off and on again(yes/no)?"
user_response = input(user_prompt)

has_tried_turning_it_on_and_off_again = user_response == "yes"

if has_tried_turning_it_on_and_off_again:
    print("Well this never happens...")
else:
    print("Turn the computer off and on again.")
````

# S'more Validator

````python
print("Welcome to the S'More Validator!")
    
pre_prompt = "Does your S'More contain "

smore_contains_graham = input(pre_prompt + "graham cracker?") == "yes"
smore_contains_marshmallow = input(pre_prompt + "marshmallow?") == "yes"
smore_contains_chocolate = input(pre_prompt + "chocolate?") == "yes"

if (
    smore_contains_graham 
    and smore_contains_marshmallow 
    and smore_contains_chocolate
):
    print("Great Looking S'More!")
else:
    print("That's not quite a S'More")
````
