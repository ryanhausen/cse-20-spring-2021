# Iterator
- An iterator is an object representing a *stream* of data
- Iterators return a single value when the `next` function is applied
- Collections like `list` can be converted into an iterable by using the built-in `iter` function
- `for` loops convert collections into an iterable when iterate over the collection
- Once the iterator has been exhausted it raises a `StopIteration` error

````python
a_list = [1, 2, 3, 4]
#print(next(a_list)) # doesn't workcd

an_iter = iter(a_list)

print(next(an_iter)) # works
print(next(an_iter)) # works
````

# Generators

- Generator functions are functions that return a *generator iterator*
- Generator functions are written in the same way as regular functions, but use `yield` instead of `return`
- Generators can also be made using a *generator expression*

# Generator Function

````python
def even_numbers():
    i = 0
    while True:
        yield i
        i += 2
        
evens = even_numbers()
for even in evens:
    print(even, end=" ")
    if even > 50:
        break
````

# Generator Function: `range`

````python
def my_range(start, stop, step=1):
    while start < stop:
        yield start
        start += step

rng = my_range(0, 10)
print(rng)
for i in rng:
    print(i, end=" ")
````

# Generator Expression

- The *generator expression* has similar syntax as the list comprehension, but rather than square brackets `[]` is uses parenthesis `()`
```python
gen = (loop_var for loop_var in collection if boolean_expression)
```

````python
evens = (i for i in range(50) if i % 2 == 0)
print(evens)
print(next(evens))
print(next(evens))
````

# `map()`

- `map` is a built-in function that takes at least two arguments function and at least one iterable(iterator/generator/collection) 
- `map` applies the function to each element in the iterable(s)
- If more than one iterable is passed, then the function must take more that number of arguments


```python
map_var = map(function, iterable1, iterable2, ....)

```

# `map()`

````python
squared = map(
    lambda x: x**2,
    [1, 2, 3, 4]
)

print(squared)
print(next(squared))
print(next(squared))
````

# `map()`

````python
squared_and_lower = map(
    lambda x, y: (x**2, y.lower()),
    [1, 2, 3],
    ["APPLE", "banana", "CaRRoT"]
)

print(squared_and_lower)
print(next(squared_and_lower))
print(next(squared_and_lower))
````

# `filter()`

- `filter` is a built-in function that takes two arguments a function and an iterable(iterator/generator/collection).
- The function argument should be a function that returns `True` or `False`
- `fitler` returns elements that when return `True` when the function argument is applied to that element

```python
filter_var = filter(function, iterable)
```

# `filter()`


````python
even_numbers = filter(
    lambda x: x % 2 == 0,
    [1, 2, 3, 4, 5]
)

print(even_numbers)
print(next(even_numbers))
print(next(even_numbers))
````

````python
def predicate(x):
    for i in range(10, 0, -1): # change to 20 for a full run
        if not x % i == 0:
            return False
    return True

def counter():
    i = 1
    while True:
        print(i, end="\r")
        yield i
        i += 1
    
value = filter(predicate, counter())
print(value)
print(next(value))
````
